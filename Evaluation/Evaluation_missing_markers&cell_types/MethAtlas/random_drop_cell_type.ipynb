```python
# Run uniform distribution deconv
from concurrent.futures import ThreadPoolExecutor
import subprocess
import numpy as np
import os

median_range = [15]

def run_cmd(cmd):
  subprocess.run(cmd, shell=True)
  
with ThreadPoolExecutor(max_workers=5) as executor:

    for median in median_range:
        ref_median_dir = f'/cfDNA_benchmark/meth_atlas_data/ref_median_{median}/drop'
        for bed_file in os.listdir(ref_median_dir):

            if bed_file.endswith('.csv'):
                bed_path = os.path.join(ref_median_dir, bed_file)
                name = bed_path.split('.')[0]
                cmd = f"python /MethAtlas_deconv.py -a {bed_path} /cfDNA_benchmark/meth_atlas_data/uniform_dis/uniform_mix/full_atlas_uniform_median_15.csv -o /cfDNA_benchmark/meth_atlas_data/uniform_dis/result/drop_ct/{name}_output.csv"

                executor.submit(run_cmd, cmd)    

import pandas as pd
import os

df = pd.read_csv('/cfDNA_benchmark/meth_atlas_data/uniform_dis/uniform_mix/full_atlas_uniform_median_15.csv')

ref_median_dir = f'/cfDNA_benchmark/meth_atlas_data/ref_median_15/drop'
for bed_file in os.listdir(ref_median_dir):
    if bed_file.endswith('.csv'):
        name = bed_file.split('.')[0]
        df.to_csv(f'/cfDNA_benchmark/meth_atlas_data/uniform_dis/uniform_mix/{name}_uniform_median_15.csv',index = False)    
```
