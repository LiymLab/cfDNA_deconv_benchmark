```python
# Generate the Constrained Random Distribution.

import random
import pandas as pd

all_list = ['Adipocytes', 'Bladder-Ep','Breast-Basal-Ep', 'Blood-B', 'Blood-Granul', 'Blood-Mono+Macro', 'Blood-NK', 'Blood-T',
            'Breast-Luminal-Ep', 'Colon-Ep', 'Colon-Fibro',
            'Endothelium', 'Eryth-prog', 'Fallopian-Ep', 'Gastric-Ep',
            'Head-Neck-Ep', 'Heart-Cardio', 'Heart-Fibro', 'Kidney-Ep',
            'Liver-Hep', 'Lung-Ep-Alveo', 'Lung-Ep-Bron', 'Neuron', 
            'Oligodend', 'Ovary+Endom-Ep', 'Pancreas-Acinar', 'Pancreas-Alpha',
            'Pancreas-Beta', 'Pancreas-Delta', 'Pancreas-Duct', 'Prostate-Ep',
            'Skeletal-Musc', 'Small-Int-Ep', 'Smooth-Musc', 'Thyroid-Ep']

blood_list = ['Blood-B', 'Blood-Granul', 'Blood-Mono+Macro', 'Blood-NK', 'Blood-T']

other_cell_list = [c for c in all_list if c not in blood_list]

df = pd.DataFrame()
df['ID'] = all_list

for i in range(100):
    n = random.randint(1, 10)
    selected = random.sample(other_cell_list, n)
    nums = [random.random() for _ in range(n+5)]
    top5 = sorted(nums, reverse=True)[:5]
    remain = sorted(nums, reverse=True)[5:]
    random.shuffle(blood_list)
    blood_dict = {k:round(v, 6) for k,v in zip(blood_list, top5)}
    cell_dict = {k:round(v, 6) for k,v in zip(selected, remain)} 
    total_dict = {**blood_dict, **cell_dict}
    
    for k in all_list:
        if k not in total_dict:
            total_dict[k] = 0
            
    total = round(sum(total_dict.values()), 6)
    for k in total_dict:
        total_dict[k] = round(total_dict[k]/total, 6)
    
    if sum(total_dict.values()) != 1:
        diff = 1 - sum(total_dict.values())
        total_dict[random.choice(blood_list)] += diff
        
    col_name = 'D'+str(i+1)
    df[col_name] = df['ID'].map(total_dict)
df_sorted = df.sort_values(df.columns[0])        
print(df)
df_sorted.to_csv('crd_dis.csv', index=False, float_format='%.6f')
```

```python
# Generate the uniform distribution.

import numpy as np
import pandas as pd

# Original data list
all_list = ['Adipocytes', 'Bladder-Ep', 'Breast-Basal-Ep', 'Blood-B', 'Blood-Granul', 'Blood-Mono+Macro', 'Blood-NK', 'Blood-T',
            'Breast-Luminal-Ep', 'Colon-Ep', 'Colon-Fibro',
            'Endothelium', 'Eryth-prog', 'Fallopian-Ep', 'Gastric-Ep',
            'Head-Neck-Ep', 'Heart-Cardio', 'Heart-Fibro', 'Kidney-Ep',
            'Liver-Hep', 'Lung-Ep-Alveo', 'Lung-Ep-Bron', 'Neuron', 
            'Oligodend', 'Ovary+Endom-Ep', 'Pancreas-Acinar', 'Pancreas-Alpha',
            'Pancreas-Beta', 'Pancreas-Delta', 'Pancreas-Duct', 'Prostate-Ep',
            'Skeletal-Musc', 'Small-Int-Ep', 'Smooth-Musc', 'Thyroid-Ep']
# Set the number of rows and columns for the matrix
rows = 35
cols = 100

# Generate a uniform distribution matrix and normalize it
uniform_matrix = np.random.uniform(size=(rows, cols))
uniform_matrix /= np.sum(uniform_matrix, axis=0)
uniform_matrix = np.round(uniform_matrix, decimals=6)

# Ensure that the sum of all columns is strictly equal to 1
column_sums = np.sum(uniform_matrix, axis=0)
for i in range(cols):
    if column_sums[i] != 1:
        diff = 1 - column_sums[i]
        random_index = np.random.choice(rows)
        uniform_matrix[random_index][i] += diff

# Create a new DataFrame with an ID column
df_with_ids = pd.DataFrame(index=np.arange(rows), columns=['ID'] + [f"D{i}" for i in range(1, cols+1)])
df_with_ids['ID'] = all_list  # Populate the ID column with elements from all_list

# Convert the uniform_matrix into a DataFrame and append it to the newly created df_with_ids
uniform_df = pd.DataFrame(uniform_matrix, columns=[f"D{i}" for i in range(1, cols+1)])
df_with_ids.iloc[:, 1:] = uniform_df

# Print the result
print(df_with_ids)

# Save the results as a CSV file
df_with_ids.to_csv('uniform_matrix.csv', index=False)
```
